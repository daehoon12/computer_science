   # 3. Analytic Geometry  
   ## Manthattan Norm ($1_1$ norm)  
   ### $ ||x||_1 = \Sigma_{i=1}^{n} |x_i|$  
   ### $ ||x||_2 = \sqrt{ \Sigma_{i=1}^{n} {x_i}^2}$
   
   ###  $ ||x||_2 = \sqrt{{x}^2 + {y}^2} = 1$
   ###  $ ||x||_1 = |y| = 1 - |x| $

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10, 10) # plt 사이즈 조절
```

```python
# 1. L2 Norm

xRight = np.linspace(0,1,50) # 0 ~ 1 사이의 값을 50등분함. [0,0.02,0.04...]
xLeft = np.linspace(-1,0,50) # -1 ~ 0 사이의 값을 50등분함.

xarr = [xRight, xLeft, xLeft, xRight]
xarr = np.array(xarr) # (4, 50)
xarr = xarr.reshape(-1) # 한줄로
# 각각 1사분면 2사분면 3사분면 4사분면 (Q1, Q2, Q3, Q4)
yarr = [np.sqrt(1-xRight**2), np.sqrt(1-xLeft**2), -np.sqrt(1-xLeft**2), -np.sqrt(1-xRight**2)]
yarr=np.array(yarr).reshape(-1)
plt.scatter(xarr, yarr, s=.5, color='r')
# 2. L1 Norm  
xarr = [xRight, xLeft, xLeft, xRight]
xarr = np.array(xarr).reshape(-1)

yarr = [1-xRight, 1+xLeft, -(1+ xLeft), -(1-xRight)]
yarr = np.array(yarr).reshape(-1)

plt.scatter(xarr, yarr, s=.5, color='b')
plt.title('Manhattan Norm (L1, blue), Euclidean Norm(L2, red)')
plt.legend(["l2", "l1"])
plt.show()
```
